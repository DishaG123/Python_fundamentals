```python
def evenoddcount(lst):
    even = 0
    odd = 0
    
    for i in lst:
        if(int(i)%2==0):
            even+=1
        else:
            odd+=1
    return even, odd
```

```python
list_4 = list()
n = int(input("Enter number of elements : "))
for i in range(0,n):
    list_4.append(int(input()))

even, odd = evenoddcount(list_4)
print('even: {} and odd :{}'.format(even,odd))
print(list_4)
```

```python
def fib(n):
    list_5 = list()
    list_5 = [0,1]
    for i in range(2,n):
        list_5.append(list_5[i-1]+list_5[i-2])

    print(list_5)
```

```python
fib(10)
```

```python
def factorial(n):
    ans = 1
    for i in range(1,n+1):
        ans *= i
    print(ans)
```

```python
factorial(5)
```

**Recursion**

```python
def rfactorial(n):
    if (n>1):
        return ( n * rfactorial(n-1) )
    else:
        return 1
```

```python
rfactorial(5)
```

**Decorators**

Used to add features to already existing funtions without altering the function itself


```python
def div (a,b):
    print(a/b)
    
def smart_div(func):
    def inner(a,b):
        if a<b :
            a,b =b,a
        return func(a,b)
    return inner
div1 = smart_div(div)
div1(2,4)
```

**Inheritance**

```python
class A:
    def feature1(self):
        print("feature_1 working")
    def feature2(self):
        print("feature_2 working")
        
class B (A):#B is child class of A
    def feature3(self):
        print("feature_3 working")
    def feature4(self):
        print("feature_4 working")
        
a1 = A()

a1.feature1()
a1.feature2()

b1 = B()
    
b1.feature1()## defined in A and inherited by B    
```
